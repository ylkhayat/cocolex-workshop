```python
methods = ["Regular", "CAD", "AdaCAD", "CoLex", "CoCoLex"]
columns = ["Correctness (RL)", "Correctness (AS)", "Faithfulness (AS)", "Fluency", "Coherence"]
def calculate_percentages(values):
    df = pd.DataFrame(values, columns=columns, index=methods)
    percentage_increase_adacad = ((df.loc["AdaCAD"] - df.loc["Regular"]) / df.loc["Regular"]) * 100
    percentage_increase_cocolex = ((df.loc["CoCoLex"] - df.loc["Regular"]) / df.loc["Regular"]) * 100
    print("AdaCAD:")
    print('-' * 10)
    print(percentage_increase_adacad)
    print("===============================")
    print("CoCoLex")
    print('-' * 10)
    print(percentage_increase_cocolex)
    print("\n\n")
```
